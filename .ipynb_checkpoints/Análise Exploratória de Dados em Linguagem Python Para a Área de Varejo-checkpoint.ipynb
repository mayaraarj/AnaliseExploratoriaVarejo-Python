## Pergunta de Negócio 1:

### Qual Cidade com Maior Valor de Venda de Produtos da Categoria 'Office Supplies'?

```python
categoria = 'Office Supplies'

#Filtrando o dataframe para incluir apenas a Categoria Office Supplies
df_office_supplies = df_dsa[df_dsa['Categoria']== categoria]

#Calculando a soma das vendas para cada cidade e agrupando o resultado por cidade
total_vendas_por_cidade = df_office_supplies.groupby('Cidade')['Valor_Venda'].sum()

# Identificando a cidade com o maior valor de vendas
cidade_maior_valor_venda = total_vendas_por_cidade.idxmax()

# Identificando o maior valor de venda
maior_valor_venda = total_vendas_por_cidade.max()

print(f"A cidade com o maior valor de venda na categoria {categoria} é {cidade_maior_valor_venda} com o total de {maior_valor_venda}")
```

## Pergunta de Negócio 2:

### Qual o Total de Vendas Por Data do Pedido?

Demonstre o resultado através de um gráfico de barras.

```python
# Calculando o total de vendas e agrupando por data do pedido
df_dsa2 = df_dsa.groupby('Data_Pedido')['Valor_Venda'].sum()

#Obs: Não converti o objeto do tipo categórico para datetime porque a pergunta não incluiu uma análise ao longo do tempo

#Gerando o gráfico
plt.figure(figsize =(20,6))
df_dsa2.plot(x='Data_Pedido', y='Valor_Venda')
plt.title('Total de Vendas por Data do Pedido')
plt.show()
```

## Pergunta de Negócio 3:

### Qual o Total de Vendas por Estado?

Demonstre o resultado através de um gráfico de barras.

```python
# Calculando o total de vendas e agrupando por data do pedido
# Reset index serve para reordenar o índice cada vez que criar uma fatia de um dataframe
df_dsa3 = df_dsa.groupby('Estado')['Valor_Venda'].sum().reset_index()

#Gerando o gráfico
plt.figure(figsize =(20,6))
sns.barplot(data= df_dsa3,
           x='Estado',
           y='Valor_Venda').set(title = 'Vendas por Estado')
plt.xticks(rotation = 70)
plt.show()
```

## Pergunta de Negócio 4:

### Quais São as 10 Cidades com Maior Total de Vendas?

Demonstre o resultado através de um gráfico de barras.

```python
# Iniciei a análise utilizando a sintaxe SQL tradicional para formular o raciocínio
# SELECT Cidade, SUM(Valor_Venda) as TotalVendas
# FROM df_dsa
# GROUP BY 'Cidade'
# ORDER BY TotalVendas DESC
# LIMIT 10

#Dividi o raciocínio em pandas atribuindo a variáveis preparatórias (mas depois vi que dava pra ir direto pra definição do dataframe)
prep1 = df_dsa.groupby('Cidade')['Valor_Venda'].sum().reset_index()
prep2 = prep1.sort_values(by='Valor_Venda', ascending = False)

#Definição do dataframe utilizado no gráfico
df_dsa4 = prep2.head(10)

#Gerando o gráfico
plt.figure(figsize =(15,5))
sns.barplot(data= df_dsa4,
           x='Cidade',
           y='Valor_Venda').set(title = '10 Cidades com Maior Total de Vendas')
plt.xticks(rotation = 70)
plt.show()
```

## Pergunta de Negócio 5:

### Qual Segmento Teve o Maior Total de Vendas?

Demonstre o resultado através de um gráfico de pizza.

```python
df_dsa5 = df_dsa.groupby('Segmento')['Valor_Venda'].sum().reset_index().sort_values(by='Valor_Venda', ascending = False)

df_dsa5.head()

#Função para converter o Valor_Venda em valor absoluto, pois etá em notação científica
def autopct_format(values):
    def my_format(pct):
        total = sum(values)
        val = int(round(pct* total/100.0))
        return ' $ {v:d}'.format(v = val)
    return my_format

#Plotando o gráfico
plt.figure(figsize=(15,5))
plt.pie(df_dsa5['Valor_Venda'],
       labels = df_dsa5['Segmento'],
       autopct = autopct_format(df_dsa5['Valor_Venda']),
       startangle = 90)

```

## Pergunta de Negócio 7 (Desafio Nível Júnior):

Os gestores da empresa estão considerando conceder diferentes faixas de descontos e gostariam de fazer uma simulação com base na regra abaixo:

- Se o Valor_Venda for maior que 1000 recebe 15% de desconto.
- Se o Valor_Venda for menor que 1000 recebe 10% de desconto.

### Quantas Vendas Receberiam 15% de Desconto?

```python
df_dsa.tail()
```

```python
#Cria coluna que verifica o valor da venda e atribui desconto
df_dsa['Desconto'] = df_dsa['Valor_Venda'].apply(lambda x: 0.15 if x >1000 else 0.10)
```

```python
#Conta e exibe quantas vendas recebeberiam 15% de desconto
contagem_desconto_15 = df_dsa['Desconto'].value_counts()
contagem_desconto_15
```

## Pergunta de Negócio 8 (Desafio Nível Master):

### Considere Que a Empresa Decida Conceder o Desconto de 15% do Item Anterior. Qual Seria a Média do Valor de Venda Antes e Depois do Desconto?

```python
df_dsa.head()
```

```python
#Cria coluna com o desconto aplicado
df_dsa['Valor_Venda_Com_Desconto'] = df_dsa['Valor_Venda']*(1-df_dsa['Desconto'])
```

```python
df_dsa.head()
```

```python
#Filtra as vendas antes do desconto de 15%
vendas_antes_desconto = df_dsa.loc[df_dsa['Desconto'] == 0.15, 'Valor_Venda']
```

```python
#Fitlra as vendas depois do desconto de 15%
vendas_depois_desconto = df_dsa.loc[df_dsa['Desconto'] == 0.15, 'Valor_Venda_Com_Desconto']
```

```python
#Calcula e imprime média do valor de venda antes do desconto de 15%
media_valor_venda_sem_desconto = vendas_antes_desconto.mean().round(2)
media_valor_venda_sem_desconto 
```

```python
#Calcula e imprime média do valor de venda depois do desconto de 15%
media_valor_venda_com_desconto = vendas_depois_desconto.mean().round(2)
media_valor_venda_com_desconto
```

## Pergunta de Negócio 9 (Desafio Nível Master Ninja):

### Qual o Média de Vendas Por Segmento, Por Ano e Por Mês?

Demonstre o resultado através de gráfico de linha.

```python
#Extrai o mês e guarda em nova variável
df_dsa['Mes'] = df_dsa['Data_Pedido'].dt.month
```

```python
#Agrupa a média de vendas por segmento, ano e mês e calcula estatísticas de agregação
df_agregacao = df_dsa.groupby(['Segmento','Ano_Pedido','Mes'])['Valor_Venda'].agg([np.sum, np.mean, np.median])
```

```python
df_agregacao
```

```python
# Extrai níveis
segmentos = df_agregacao.index.get_level_values(0)
anos = df_agregacao.index.get_level_values(1)
meses = df_agregacao.index.get_level_values(2)
```

```python
#Plota o gráfico
plt.figure(figsize =(12,6))

#Limpa a formatação do seaborn
sns.set()

#Cria gráfico de relação entre as variáveis
fig1 = sns.relplot(kind = 'line',
                   data = df_agregacao, 
                   y = 'mean', 
                   x = meses,
                   hue = segmentos, 
                   col = anos,
                   col_wrap = 4)
plt.show()
```

## Pergunta de Negócio 10 (Desafio Nível Master Ninja das Galáxias):

### Qual o Total de Vendas Por Categoria e SubCategoria, Considerando Somente as Top 12 SubCategorias? 

Demonstre tudo através de um único gráfico.

```python
# Agrupamos por categoria e subcategoria e calculamos a soma somente para variáveus numéricas
df_dsa_p10 = df_dsa.groupby(['Categoria',
                             'SubCategoria']).sum(numeric_only = True).sort_values('Valor_Venda',
                                                                                   ascending = False).head(12)
```

```python
# Convertemos a coluna Valor_Venda em número inteiro e classificamos por categoria
df_dsa_p10 = df_dsa_p10[['Valor_Venda']].astype(int).sort_values(by = 'Categoria').reset_index()
df_dsa_p10
```

```python
# Criamos outro dataframe somente com os totais por categoria
df_dsa_p10_cat = df_dsa_p10.groupby('Categoria').sum(numeric_only = True).reset_index()
df_dsa_p10_cat
```

```python
# Listas de cores para categorias
cores_categorias = ['#5d00de',
                    '#0ee84f',
                    '#e80e27']
```

```python
# Listas de cores para subcategorias
cores_subcategorias = ['#aa8cd4',
                       '#aa8cd5',
                       '#aa8cd6',
                       '#aa8cd7',
                       '#26c957',
                       '#26c958',
                       '#26c959',
                       '#26c960',
                       '#e65e65',
                       '#e65e66',
                       '#e65e67',
                       '#e65e68']
```

```python
# Plot

# Tamanho da figura
fig, ax = plt.subplots(figsize = (18,12))

# Gráfico das categorias
p1 = ax.pie(df_dsa_p10_cat['Valor_Venda'], 
            radius = 1,
            labels = df_dsa_p10_cat['Categoria'],
            wedgeprops = dict(edgecolor = 'white'),
            colors = cores_categorias)

# Gráfico das subcategorias
p2 = ax.pie(df_dsa_p10['Valor_Venda'],
            radius = 0.9,
            labels = df_dsa_p10['SubCategoria'],
            autopct = autopct_format(df_dsa_p10['Valor_Venda']),
            colors = cores_subcategorias, 
            labeldistance = 0.7,
            wedgeprops = dict(edgecolor = 'white'), 
            pctdistance = 0.53,
            rotatelabels = True)

# Limpa o centro do círculo
centre_circle = plt.Circle((0, 0), 0.6, fc = 'white')

# Labels e anotações
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.annotate(text = 'Total de Vendas: ' + '$ ' + str(int(sum(df_dsa_p10['Valor_Venda']))), xy = (-0.2, 0))
plt.title('Total de Vendas Por Categoria e Top 12 SubCategorias')
plt.show()
```
